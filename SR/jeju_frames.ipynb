```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.windows import Window
from PIL import Image
import numpy as np

image_path = '/content/drive/MyDrive/jeju_png/clipping/rgb_image_8.tif'
output_image_path = '/content/drive/MyDrive/lr_fin/08신양섭지코지해변.jpg'

pixel_x, pixel_y =  2.123e+04, 1.474e+04
zoom_factor = 2  # 확대 비율 (2배 확대)

# 이미지 열기
with rasterio.open(image_path) as src:
    # 원본 윈도우 크기 설정 (예: 640x640)
    original_window_size = 640

    # 줌 윈도우 크기 계산 (더 작은 영역을 선택하기 위해 확대 비율로 나눔)
    zoom_window_size = original_window_size // zoom_factor

    # 윈도우 설정 (더 작은 영역을 선택)
    window = Window(pixel_x - zoom_window_size // 2, pixel_y - zoom_window_size // 2, zoom_window_size, zoom_window_size)

    # 선택된 윈도우로 이미지 데이터 읽기
    data = src.read(window=window)

    # 데이터가 3밴드(RGB)인지 확인하고 처리
    if data.shape[0] == 3:
        img_array = np.transpose(data, (1, 2, 0))
        img = Image.fromarray(img_array.astype('uint8'), 'RGB')
    else:
        img = Image.fromarray(data[0].astype('uint8'), 'L')

    # 이미지를 JPEG로 저장
    img.save(output_image_path, 'JPEG')

    # 선택된 작은 영역을 원래 크기로 리샘플링 (확대)
    img = img.resize((original_window_size, original_window_size), Image.LANCZOS)

    # 이미지를 JPEG로 저장
    img.save(output_image_path, 'JPEG')

    # 결과 이미지 시각화
    plt.imshow(img)
    plt.title("Zoomed Image at Pixel Location")
    plt.colorbar()
    plt.show()
```

```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.windows import Window
from PIL import Image
import numpy as np

# 경로 설정
image_path = '/content/drive/MyDrive/jeju_png/clipping/rgb_image_8.tif'
output_image_path = '/content/drive/MyDrive/lr_fin/08신양섭지코지해변2.jpg'

# 첫 번째 자르기의 중심 좌표와 크기 (원본 이미지 기준)
initial_center_x, initial_center_y = 2.123e+04, 1.474e+04
initial_window_size = 640  # 첫 번째 자른 이미지의 크기

# 확대할 부분의 중심 좌표 (첫 번째 자른 이미지 내 상대 좌표)
zoom_center_x, zoom_center_y = 220, 460  # 첫 번째 자른 이미지 내에서의 중심 좌표

zoom_factor = 1  # 확대 비율 (예: 4배 확대)

# 이미지 열기
with rasterio.open(image_path) as src:
    # 첫 번째 자르기: 원본 이미지에서 640x640 크기로 자름
    initial_window = Window(initial_center_x - initial_window_size // 2,
                            initial_center_y - initial_window_size // 2,
                            initial_window_size, initial_window_size)

    # 첫 번째 자른 이미지 데이터 읽기
    initial_data = src.read(window=initial_window)

    # 두 번째 자르기: 첫 번째 자른 이미지 내에서 확대할 영역 선택
    zoom_window_size = initial_window_size // zoom_factor  # 확대할 작은 영역의 크기 계산

    # 확대할 영역이 첫 번째 자른 이미지 내에서 벗어나지 않도록 좌표를 확인
    zoom_center_x = max(zoom_center_x, zoom_window_size // 2)
    zoom_center_x = min(zoom_center_x, initial_window_size - zoom_window_size // 2)

    zoom_center_y = max(zoom_center_y, zoom_window_size // 2)
    zoom_center_y = min(zoom_center_y, initial_window_size - zoom_window_size // 2)

    # zoom_center_x, zoom_center_y는 자른 이미지 내에서의 좌표이므로, 이 값을 기준으로 새로운 윈도우 설정
    zoom_window = Window(zoom_center_x - zoom_window_size // 2,
                         zoom_center_y - zoom_window_size // 2,
                         zoom_window_size, zoom_window_size)

    # 두 번째 윈도우에 맞게 데이터 자르기
    zoomed_data = initial_data[:,
                               int(zoom_window.col_off):int(zoom_window.col_off + zoom_window.width),
                               int(zoom_window.row_off):int(zoom_window.row_off + zoom_window.height)]

    # 자른 이미지를 원래 크기인 640x640으로 리샘플링 (확대)
    img_array = np.transpose(zoomed_data, (1, 2, 0))
    zoomed_img = Image.fromarray(img_array.astype('uint8'), 'RGB')
    zoomed_img = zoomed_img.resize((initial_window_size, initial_window_size), Image.LANCZOS)

    # # 이미지를 JPEG로 저장
    # zoomed_img.save(output_image_path, 'JPEG')

    # 결과 이미지 시각화
    plt.imshow(zoomed_img)
    plt.title("Zoomed Image at Specific Location")
    plt.colorbar()
    plt.show()
```
