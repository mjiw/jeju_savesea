고화질, 저화질 이미지쌍만 남겨두기

```python
import os

def sync_folders(gt_folder, lq_folder):
    # Get list of image filenames in both folders
    gt_images = set(os.listdir(gt_folder))
    lq_images = set(os.listdir(lq_folder))

    # Find images that are in gt but not in lq
    gt_only = gt_images - lq_images
    # Find images that are in lq but not in gt
    lq_only = lq_images - gt_images

    # Delete images that are only in gt folder
    for img_name in gt_only:
        img_path = os.path.join(gt_folder, img_name)
        print(f"Deleting {img_path}")
        os.remove(img_path)

    # Delete images that are only in lq folder
    for img_name in lq_only:
        img_path = os.path.join(lq_folder, img_name)
        print(f"Deleting {img_path}")
        os.remove(img_path)

# Example usage
gt_folder = '/content/drive/MyDrive/train/hr_sub'  # 고화질 이미지 폴더 경로
lq_folder = '/content/drive/MyDrive/train/low'  # 저화질 이미지 폴더 경로

sync_folders(gt_folder, lq_folder)
```

평가데이터 작게 자르기

```python
import cv2
import os

# 이미지가 있는 디렉토리 경로 설정
input_dir = '/content/drive/MyDrive/train/hr_sub'
output_dir = '/content/drive/MyDrive/train/lr_sub'

# 출력 디렉토리가 존재하지 않으면 생성
os.makedirs(output_dir, exist_ok=True)

# 이미지 파일 확장자 목록
img_extensions = ['.jpg', '.jpeg', '.png', '.bmp']

# 디렉토리 내 모든 파일에 대해 작업 수행
for filename in os.listdir(input_dir):
    if any(filename.lower().endswith(ext) for ext in img_extensions):
        img_path = os.path.join(input_dir, filename)

        # 이미지 로드
        img = cv2.imread(img_path)

        # 이미지가 정상적으로 로드되었는지 확인
        if img is None:
            print(f"Failed to load image {img_path}")
            continue

        # 이미지의 높이와 너비 구하기
        height, width, _ = img.shape

        # 4등분한 이미지의 크기 계산
        half_height = height // 2
        half_width = width // 2

        # 파일 이름에서 확장자 제거
        base_name = os.path.splitext(filename)[0]

        # 1번 이미지 (왼쪽 상단)
        img1 = img[:half_height, :half_width]
        cv2.imwrite(os.path.join(output_dir, f'{base_name}_part1.jpg'), img1)

        # 2번 이미지 (오른쪽 상단)
        img2 = img[:half_height, half_width:]
        cv2.imwrite(os.path.join(output_dir, f'{base_name}_part2.jpg'), img2)

        # 3번 이미지 (왼쪽 하단)
        img3 = img[half_height:, :half_width]
        cv2.imwrite(os.path.join(output_dir, f'{base_name}_part3.jpg'), img3)

        # 4번 이미지 (오른쪽 하단)
        img4 = img[half_height:, half_width:]
        cv2.imwrite(os.path.join(output_dir, f'{base_name}_part4.jpg'), img4)

        print(f"Processed {filename} and saved parts to {output_dir}")

print("All images have been processed and saved.")
```

```python
import os

def count_images_in_directory(directory_path):
    # 이미지 파일의 확장자 목록 (필요에 따라 추가 가능)
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp']

    # 이미지 파일 개수를 세기 위한 변수
    image_count = 0

    # 디렉토리 내의 모든 파일 및 폴더 탐색
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            # 파일의 확장자가 이미지 확장자 목록에 포함되어 있는지 확인
            if os.path.splitext(file)[1].lower() in image_extensions:
                image_count += 1

    return image_count

# 예시 사용
directory_path = '/content/drive/MyDrive/train/lr_sub'
print(f"디렉토리 내 이미지 파일 개수: {count_images_in_directory(directory_path)}")

# /content/drive/MyDrive/train_HR 800
# /content/drive/MyDrive/valid_HR 64
```

```python
import cv2
import os

def check_image_files_in_folder(folder_path):
    if not os.path.exists(folder_path):
        print(f"Folder does not exist at path: {folder_path}")
        return

    # 리스트에 모든 파일을 가져옵니다.
    files = os.listdir(folder_path)

    # PNG 파일만 필터링합니다.
    png_files = [f for f in files if f.lower().endswith('.png')]

    if not png_files:
        print("No PNG files found in the folder.")
        return

    for file_name in png_files:
        file_path = os.path.join(folder_path, file_name)
        img = cv2.imread(file_path)
        if img is None:
            print(f"Failed to load image at path: {file_path}")
        else:
            print(f"Image loaded successfully at path: {file_path}")

# 예시 사용법
folder_path = '/content/drive/MyDrive/valid/lr_sub'  # PNG 파일이 저장된 폴더 경로
check_image_files_in_folder(folder_path)
```

```python
import os
from PIL import Image

def create_metadata_txt(image_dir, output_txt):
    with open(output_txt, 'w') as file:
        for image_name in os.listdir(image_dir):
            if image_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_path = os.path.join(image_dir, image_name)
                with Image.open(image_path) as img:
                    width, height = img.size
                    channels = len(img.getbands())
                    file.write(f"{image_name} ({width},{height},{channels})\n")

# 사용 예제
image_dir = '/content/drive/MyDrive/valid/hr_sub'  # 이미지가 저장된 디렉토리 경로
output_txt = '/content/drive/MyDrive/valid/metadata.txt'        # 출력할 텍스트 파일 경로
create_metadata_txt(image_dir, output_txt)
```
